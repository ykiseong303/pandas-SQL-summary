```python
# SQL은 연산에 굉장히 최적화가 되어 있어서 빠르게 가능!

# 100만개 => name, population => 네트워크에 대한 비용 => df
# 비용에 대해 좀 오래걸림. COUNT를 하면 서버에서 하고 보내는거니 연산이 빠름

```

```python
# Continent == Europe, Population > 10,000,000 데이터 뽑아오기
```

```python
# pandas
is_europe = country_df["Continent"] == "Europe"
is_population = country_df["Population"] > 10000000

country_df[is_europe][is_population].head()

# and : 유럽에 있고, 인구가 천만 이상인 애들
# country_df[is_europe & is_population]
# or : 유럽에 있거나 인구가 천만 이상인 애들
# country_df[is_europe | is_population]
```

```python
# SQL
SQL_QUERY = """
    SELECT *
    FROM Country
    WHERE 
        Continent = "Europe" 
        AND Population > 10000000
    ;
"""

pd.read_sql(SQL_QUERY, db).head()
```
